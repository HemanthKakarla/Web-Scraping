```python
import requests
import bs4
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import lxml # Optional ; to understand html code
```

```python
data = requests.get('https://www.formula1.com/en/results.html/2019/races/1000/australia/race-result.html')
data
```

```python
data_bs = BeautifulSoup(data.text)
data_bs
```

```python
Driver=[]
Team=[]
Position=[]
Track=[]
```

### Drivers


```python
data_tag_driver=data_bs.find_all('span',class_=['hide-for-mobile','hide-for-tablet'])
data_tag_driver
for i in range(0,40,2):
    name1 = data_tag_driver[i].text
    name2 = data_tag_driver[i+1].text
    name=name1+' '+name2
    Driver.append(''.join(name))
Driver
```

### Positions

```python
data_tag_pos=data_bs.find_all('td',class_=['dark'])
text=''
for i in data_tag_pos:
    pos = i.text
    
    text=" ".join((text,pos))
pos = re.findall('(\d+)\s\d+\s\\n',text)
pos1 = re.findall('([A-Z]+)\s\d+\s\\n',text)
```

```python
Position=pos+pos1
Position
```

### Teams

```python
data_tag_team=data_bs.find_all('td',class_=['semi-bold uppercase hide-for-tablet'])
Team=[]
for i in data_tag_team:
    team=i.text
    Team.append(''.join(team))
Team
```

```python
Track=['Australia']*20
Track
Year=['2019']*20
```

```python
df = pd.DataFrame({'Year':Year,
                    'Track':Track,
                  'Position':Position,
                  'Driver':Driver,
                  'Team':Team})
```
